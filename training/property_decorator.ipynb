```python
import math

class RockMass:
    def __init__(self, ucs, friction_angle):
        self.ucs = ucs
        #self._friction_angle = None # Initialize hidden variable, but setter will auto do it
        self.friction_angle = friction_angle # Triggers the setter! -> if pass, set as _fa

    @property # this is getter
    def friction_angle(self):
        return self._friction_angle

    @friction_angle.setter
    def friction_angle(self, value):
        if value < 0 or value > 90:
            raise ValueError(f"Friction angle {value} is physically impossible (0-90).")
        if not isinstance(value, (int, float)):
            raise TypeError("Signal interference: Friction angle must be Numeric.")
        self._friction_angle = value

    def estimated_strength(self, confinement, stress_dir=None):  # only putitng this stress dirn for compatibility with our himalayan rock

        # Mohr-Coulomb: sigma1 = ucs + k * confinement
        # (Simplified for drill purposes)
        k = (1 + math.sin(math.radians(self.friction_angle))) / (1 - math.sin(math.radians(self.friction_angle)))
        return self.ucs + (k * confinement)

    @property
    def ucs(self):
        return self._ucs

    @ucs.setter
    def ucs(self, value):
        if  not (0 <= value <= 100):
            raise ValueError(f"UCS {value} is not within 0-100 limit.")
        self._ucs = value

lime = RockMass(ucs=19, friction_angle=10)
#shale = RockMass(friction_angle=120)
```

```python
class HimalayanRock(RockMass):
    """rock found in the himalayas"""
    def __init__(self, ucs, friction_angle, foliation_dir):
        super().__init__(ucs, friction_angle)
        self.foliation_dir = foliation_dir
    # fxn to override estimated str

    def estimated_strength(self, confinement, stress_dir=None): # like bayesian updating, we are updating the est_str here using a new param called stress_dir

    # need to use here None in stress drn to maintian same operational constraints as the parents
    
        base_str = super().estimated_strength(confinement)

        if stress_dir is not None and abs(stress_dir-self.foliation_dir)<=10:
            updated_base_str = base_str*0.68
            return updated_base_str
        return base_str

```
